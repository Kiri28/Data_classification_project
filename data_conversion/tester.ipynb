### Можем попробовать использовать объединения слов для того чтобы работать со словосочетаниями и смотреть что нибудь по ним.

```python
from __future__ import unicode_literals
import nltk
from nltk import word_tokenize
from nltk.util import ngrams
from collections import Counter

token = nltk.wordpunct_tokenize(new_text)
bigrams = ngrams(token,2)
trigrams = ngrams(token,3)
fourgrams = ngrams(token,4)
fivegrams = ngrams(token,5)

for k1, k2 in Counter(bigrams):
         print (k1+"_"+k2)

for k1, k2,k3 in Counter(trigrams):
         print (k1+"_"+k2+"_"+k3)

for k1, k2,k3,k4 in Counter(fourgrams):
         print (k1+"_"+k2+"_"+k3+"_"+k4)

for k1, k2,k3,k4,k5 in Counter(fivegrams):
         print (k1+"_"+k2+"_"+k3+"_"+k4+"_"+k5)
```
