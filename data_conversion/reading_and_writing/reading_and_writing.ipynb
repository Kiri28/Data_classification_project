```python
#Функция для обработки словосочетаний

from __future__ import unicode_literals
import nltk
from nltk import word_tokenize
from nltk.util import ngrams
from collections import Counter

def tokens(new_text):
    #Массив, который будет выводить словосочетания
    ret=[]
    
    token = nltk.wordpunct_tokenize(new_text)
    bigrams = ngrams(token,2)
    trigrams = ngrams(token,3)
    fourgrams = ngrams(token,4)
    fivegrams = ngrams(token,5)

    if len(new_text.split())>=2:
        for k1, k2 in Counter(bigrams):
            ret.append(k1+"_"+k2)

    if len(new_text.split())>=3:
        for k1, k2,k3 in Counter(trigrams):
            ret.append(k1+"_"+k2+"_"+k3)

    if len(new_text.split())>=4:
        for k1, k2,k3,k4 in Counter(fourgrams):
            ret.append(k1+"_"+k2+"_"+k3+"_"+k4)

    if len(new_text.split())>=5:
        for k1, k2,k3,k4,k5 in Counter(fivegrams):
            ret.append(k1+"_"+k2+"_"+k3+"_"+k4+"_"+k5)
        
    return(ret)
```
